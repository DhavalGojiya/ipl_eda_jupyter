# Cleansing The Data

```python
data.isnull().sum()
```

# Merging the dataset

```python
merge = pd.merge(deliveries,data,left_on='match_id', right_on ='id')
merge.head()
```

```python
merge.describe()
```

```python
merge.head(5)
```

# Batsman with Highest Sixes

```python
six=merge[merge['batsman_runs']==6]
c=six.groupby('batsman').count()['batsman_runs'].sort_values(ascending=False).head(10)
c
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=merge,x=c.index,y=c)
```

# TOP WICKET TAKERS

```python
deliveries.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).plot(kind='bar',color='purple')
plt.title('Top Wicket Taker of IPL')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets Taken')
plt.show()
```
